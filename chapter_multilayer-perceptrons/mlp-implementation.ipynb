# Implementation of Multilayer Perceptrons
:label:`sec_mlp-implementation`

Multilayer perceptrons (MLPs) are not much more complex to implement than simple linear models. The key conceptual
difference is that we now concatenate multiple layers.


```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
model = MLPScratch(num_inputs=784, num_outputs=10, num_hiddens=100, lr=0.1)
data = d2l.FashionMNIST(batch_size=256)
trainer = d2l.Trainer(max_epochs=10)
trainer.fit(model, data)
```

O Modelo 2 parece ser o mais equilibrado, com uma boa diminuição na perda de treinamento e validação, e um aumento consistente na precisão de validação.

O Modelo 1 pode estar começando a dar overfitting, com a perda de validação não diminuindo na mesma taxa que a perda de treinamento.

O Modelo 3 pode estar sofrendo de overfitting, com a perda de validação aumentando ou diminuindo muito lentamente, e a precisão de validação estagnando.

Modelo 2 é o mais promissor, pois mostra um bom equilíbrio entre ajuste aos dados de treinamento e generalização para dados de validação.

```python
class MLP(d2l.Classifier):
    def __init__(self, num_outputs, num_hiddens, num_hiddens2, lr):
        super().__init__()
        self.save_hyperparameters()
        self.net = nn.Sequential(nn.Flatten(), nn.LazyLinear(num_hiddens), nn.ReLU(), nn.LazyLinear(num_hiddens2), nn.ReLU(), nn.LazyLinear(num_outputs))

model = MLP(num_outputs=10, num_hiddens=256, num_hiddens2=257, lr=0.2)
trainer.fit(model, data)
```

O Gráfico 2 representa um modelo mais equilibrado e eficiente, com melhor generalização e sem sinais claros de overfitting.

O Gráfico 1 sugere que o modelo pode estar começando a overfitar, com a perda de validação diminuindo mais lentamente e a precisão de validação se estabilizando.
